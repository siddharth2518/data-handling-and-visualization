```
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[4,5,6,7]
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Line graph")
plt.plot(x,y)
plt.show()
```

```
x=[20,30,40,50]
plt.pie(x)
plt.show()
```

```
y=[10,20,30,60]
mylabels=["Apples","bananas","cherries","Dates"]
plt.pie(y,labels = mylabels)
plt.show()
```

```
z=[10,30,40,50]
mylabels=["apple","banana","cherries","Dates"]
mycolors=["black","red","blue","#4CAF50"]
plt.pie(z,labels=mylabels,colors=mycolors)
plt.show
```

```

y=[10,20,30,50]
mylabels=["apple","banana","cherries","Dates"]
myexplode=[0,0,0,0.5]
plt.pie(y,labels=mylabels,explode=myexplode)
plt.show
```

```
#bar graph
import numpy as np
x=np.array(["A","B","c","D"])
y=np.array([3,8,1,10])
plt.bar(x,y)
plt.show()

```

```
x=np.array(["A","B","c","D"])
y=np.array([3,8,1,10])
plt.scatter(x,y)
plt.show()
```

```
x=np.array([1,2,3,4])
y=np.array([1,2,3,4])
plt.scatter(x,y,color="r")
plt.show()
```

```
x=np.array([1,2,3,4])
y=np.array([1,2,3,4])
mycolors=["r","b","y","g"]
plt.scatter(x,y,color=mycolors)
plt.show()
```

```
x=np.array([1,2,3,4])
y=np.array([1,2,3,4])
z=np.array([6,10,40,30])
plt.scatter(x,y,z,color="r")
plt.show()
```

```
x=np.array([1,2,3,4])
y=np.array([1,2,3,4])
z=np.array([6,10,40,30])
mycolors=["r","b","y","g"]
plt.scatter(x,y,z,color=mycolors)
plt.show()
```

```
x=np.array([1,2,3,4])
y=np.array([1,2,3,4])
sizes=np.array([6,10,40,30])
colors=["r","b","y","g"]
plt.scatter(x,y,c=colors,s=sizes,alpha=0.8,marker="*",edgecolor="g")
plt.title("scatter",fontsize=20)
plt.xlabel("x axis",fontsize=20)
plt.show()
```

```
plt.subplot(2,2,1)
x=[1,2,3,4]
y=[4,5,6,7]
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Line graph")
plt.plot(x,y)

plt.subplot(2,2,2)
x=np.array([1,2,3,4])
y=np.array([1,2,3,4])
z=np.array([6,10,40,30])
mycolors=["r","b","y","g"]
plt.scatter(x,y,z,color=mycolors)

plt.subplot(2,2,3)
x=np.array(["A","B","c","D"])
y=np.array([3,8,1,10])
plt.bar(x,y)

plt.subplot(2,2,4)
z=[10,30,40,50]
mylabels=["apple","banana","cherries","Dates"]
mycolors=["black","red","blue","#4CAF50"]
plt.pie(z,labels=mylabels,colors=mycolors)
plt.show
```
